# Handling Missing Data

```python
dataset.isnull().sum()
```

```python
data = dataset.drop(['Property Type', 'Residential Type', 'Non Use Code', 'Assessor Remarks', 'OPM remarks', 'Location','Sales Ratio'], axis=1)
```

```python
town=data.Town.values.reshape(-1,1)
address=data.Address.values.reshape(-1,1)
date=data['Date Recorded'].values.reshape(-1,1)
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
data.Address=label_encoder.fit_transform(data.Address)
data.Town=label_encoder.fit_transform(data.Town)
data['Date Recorded']=label_encoder.fit_transform(date)
```
